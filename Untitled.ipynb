# Culture (1)

```python
import glob
import pandas as pd
files = glob.glob("E:/كلية/4th cs/Secound Term/Arabization(تعريف لغات الحاسب)/Data set/New folder (4)/New folder/Culture/*.txt" , recursive = True)
```

```python
s1 = [] 
for file in files:
    data1 = open(file , "r" , encoding = "utf-8")
    s1.append(data1.read())
```

```python
dataframe1 = []
for i in s1:
    dataframe1.append([i ,"Culture" ])
df1 = pd.DataFrame(dataframe1 , columns = ["text" , "label"])
```

```python
df1.to_csv("Culture_topic.csv" , encoding="utf-8" , index=False)
```

```python
df=pd.read_csv("Culture_topic.csv" ,encoding="utf-8" )
```
